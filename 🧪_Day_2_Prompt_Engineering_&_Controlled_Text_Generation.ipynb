```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

def generate_response(prompt):
  output = generator(
      prompt,
      max_length=50,
      temperature=0.5,
      top_k=50,
      top_p=0.9,
      num_return_sequences=1,
      repetition_penalty=1.3,
      pad_token_id=50256
  )
  return output[0]['generated_text']

prompt = """You are a helpful AI assistant having a technical conversation with a user.

Yasir: What is the difference between AI and ML?
AI:"""
response = generate_response(prompt)
print(response)
```
