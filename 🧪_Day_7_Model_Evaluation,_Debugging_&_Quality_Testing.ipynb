```python
from transformers import pipeline

generator = pipeline('text-generation', "devxyasir/historical-gpt2-finetuned")
```

```python
def test_model(prompt):
  output = generator(
      prompt,
      max_length=100,
      do_sample=True,
      temperature=0.7,
      top_k=50,
      top_p=0.9,
      repetition_penalty=1.05,
      num_return_sequences=1,
      pad_token_id=50256
  )[0]['generated_text']
  return output

print(test_model("What role did the Muslim League play?"))
```
